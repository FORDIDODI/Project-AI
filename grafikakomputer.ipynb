```python
import matplotlib.pyplot as plt
#membuat canvas 500x500
plt.figure(figsize=(5,5))
#menampilkan titik (x,y) dalam koordinat
x = [200]
y = [150]
plt.scatter(x,y,color='red', s=50) #s = ukuran titik
plt.title("Plot Titik Pertama")
plt.xlim(0,500)
plt.ylim(0,500)
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
#Koordinat latihan mahasiswa
x = [50,120,250,350,450]
y = [60, 200, 300,400, 450]
plt.figure(figsize=(5,5))
plt.scatter(x,y,color='blue',s=80, marker='o')
plt.title("Latihan Mandiri - 5 Titik Baru")
plt.xlim(0,500)
plt.ylim(0,500)
plt.grid
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [100,200,300,400]
y = [100,150,250,350]
plt.figure(figsize=(5,5))
plt.plot(x,y,color='green', marker='o', linewidth=2)
#titik dihubungkan garis
plt.title("Menghubungkan Titik dengan garis")
plt.xlim(0,500)
plt.ylim(0.500)
plt.grid
plt.show()
```
