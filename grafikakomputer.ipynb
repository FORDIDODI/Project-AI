```python
import matplotlib.pyplot as plt

# Data koordinat (menggunakan titik desimal)
x = [4, 5, 6, 7, 8, 9, 10, 11, 12]
y = [8, 9, 10, 11, 11, 12, 13, 14, 15]  # asumsi Y8

plt.figure(figsize=(6,6))
plt.plot(x, y, color='green', marker='o')   # garis + marker
plt.title("Plot Data Koordinat X-Y")
plt.xlabel("X")
plt.ylabel("Y")
plt.xlim(3,12)   # batas agar titik terlihat jelas
plt.ylim(7,15)
plt.grid(True)   # aktifkan grid
plt.show()
```

```python
def dda(x1, y1, x2, y2) :
    x, y = x1, y1
    dx = x2 - x1
    dy = y2 - y1
    steps = int (max(abs(dx), abs(dy)))
    x_inc = dx / steps
    y_inc = dy / steps

    xs, ys = [], []
    for i in range(steps+1) :
        xs.append(round(x))
        ys.append(round(y))
        x += x_inc
        y += y_inc
    return xs, ys

x, y = dda(3, 12, 7, 15)

plt.figure (figsize=(5,5))
plt.scatter (x,y, color='red')
plt.plot (x, y, color='red', linewidth=1)
plt.title ("Algoritma DDA")
plt.xlim(0,16); plt.ylim(0, 16); 
plt.grid(True)
plt.show()
```

```python
def brensenham (x1,y1,x2,y2):
  x, y = x1,y1
  dx = x2 - x1
  dy = y2 - y1
  p = 2*dy - dx
  xs, ys = [],[]
  for i in range(x1, x2+1):
    xs.append(x)
    ys.append(y)
    x +=1
    if p >= 0:
      y += 1
      p = p + 2*dy - 2*dx
    else:
      p += 2*dy 
  return xs,ys
x, y = brensenham (2, 1, 8, 5)
plt.figure(figsize=(10,10))
plt.scatter(x, y,color='red')
plt.plot(x, y,color='blue', linewidth=1)
plt.title("Algorithm Bresenham")
plt.xlim(0,10)
plt.ylim(0,10)
plt.grid(True)
plt.show()
```

```python
def midpoint_circle(xc, yc, r) :
    r = int(round(r))
    x = 0
    y = r
    p = 1 - r
    points = []
    while x <= y :
        # 8 - simetri lingkaran
        points.extend([
            (xc + x, yc + y), (xc - x, yc + y),
            (xc + x, yc - y), (xc - x, yc - y),
            (xc + y, yc + x), (xc - y, yc + x),
            (xc + y, yc - x), (xc - y, yc - x)
        ])

        x += 1
        if p < 0 :
            p = 2*x + 1
        else :
            y -= 1
            p = p + 2*(x - y) + 1
    return points

#contoh penggunaan
xc,yc, r = 200, 200, 100
points = midpoint_circle(xc, yc, r)
x, y = zip(*points)

plt.figure(figsize=(5,5))
plt.scatter(x, y,color='blue', s=10)
plt.title("MidPoint Circle Aglgorithm")
plt.xlim(0,400)
plt.ylim(0,400)
plt.gca().set_aspect('equal')
plt.grid(True)
plt.show()
```
